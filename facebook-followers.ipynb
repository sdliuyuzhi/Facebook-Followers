```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
def read_file():
    df = pd.concat([pd.read_csv(f, parse_dates=True,
                                low_memory=False,
                                encoding = "ISO-8859-1", thousands=',') 
                              for f in glob.glob('*csv')])
    df['time'] = pd.to_datetime(df['time'])
    df['date_updated'] = pd.to_datetime(df['date_updated'])
    return df

def get_nth_likes_company(df, n, id_list):
    f_id = id_list[n - 1]
    return df[df['facebook_id'] == f_id]

def get_company_id_list(df):
    tmp = df.sort_values('likes', ascending=False)
    uniq_id = tmp['facebook_id'].unique()
    return uniq_id

def plot_a_vs_b(df, str1, str2):
    df = df.sort_values(str1, ascending=True)
    name = df['username'].iloc[0]
    df.plot(x=str1,y=str2, title=name)
    plt.show()
    return
```

```python
df = read_file()
```
